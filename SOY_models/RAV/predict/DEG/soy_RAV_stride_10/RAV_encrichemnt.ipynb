```python
import pandas as pd
import numpy as np
from io import StringIO 
import pyranges as pr
from numpy import array
import pybedtools
```

```python
fimo_pred = pd.read_csv("DEm_fimo_5999.tsv",sep="\t",header=0)
fimo_pred
```

```python
fimo_pred = fimo_pred.dropna(axis=0)
fimo_pred
```

```python
RAV_fimo=fimo_pred[fimo_pred['motif_alt_id'].str.contains("RAV")]
RAV_fimo
```

```python
fimo_pred_bed=RAV_fimo.drop(["motif_id","p-value","q-value","q-value","strand","score","matched_sequence"], axis=1)
fimo_pred_bed["start"]= (fimo_pred_bed["start"]).astype(int)
fimo_pred_bed["stop"]= (fimo_pred_bed["stop"]).astype(int)

fimo_pred_bed=fimo_pred_bed[["sequence_name","start","stop","motif_alt_id"]]
fimo_pred_bed
```

```python
fimo_motif_scan_bed_rav = pybedtools.BedTool.from_dataframe(fimo_pred_bed)\
              .saveas('fimo_pred_RAV_5999DEG.bed')
#print(fimo_motif_scan_bed)
```

```python
a=fimo_motif_scan_bed_rav
```

```python
wrky_10w_pred = pybedtools.BedTool('RAV_10W_DEG5999_target_windows_0.8thresh.bed')


b=wrky_10w_pred
b
```

```python
a_and_b = a.intersect(b,wb=True)
```

```python
WRKY_motfi_intersect = pybedtools.BedTool.to_dataframe(a_and_b)
WRKY_motfi_intersect
```

```python
WRKY_motfi_intersect.to_csv('RAV_pred_fimo_overlap_nofilt.csv')  
```
