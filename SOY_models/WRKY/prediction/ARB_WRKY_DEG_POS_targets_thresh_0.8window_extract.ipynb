```python
import pandas as pd
import numpy as np
from io import StringIO 
import pyranges as pr
```

```python
WRKY_pred = pd.read_csv("soy_WRKY_201bp_targets_DEG5999_0612_thres08_windowloc_w10.csv",header=0)
print(WRKY_pred  )
```

```python
df_WRKY_pred =WRKY_pred .drop(['n'],axis=1)
df_WRKY_pred 
```

```python
promtoer_split = pd.DataFrame(df_WRKY_pred .tgenes.str.split('::',2).tolist(),columns=['gene','location1'])

promtoer_split
```

```python
pr_loc_split = pd.DataFrame(promtoer_split.location1.str.split(':',2).tolist(),columns=['chrom','location'])
pr_loc_split
```

```python
prom_cod = pd.DataFrame(pr_loc_split.location.str.split('-',2).tolist(),columns=['start','end'])
prom_cod
```

```python
promoter_wnd=pd.concat([df_WRKY_pred ,promtoer_split,pr_loc_split,prom_cod],axis=1)
 
promoter_wnd
```

```python
df_promoter_wnd=promoter_wnd.drop(['tgenes','location1','location'], axis=1)

#convert string data type to int
df_promoter_wnd['start']=df_promoter_wnd['start'].astype(int)
df_promoter_wnd['end']=df_promoter_wnd['end'].astype(int)
print(df_promoter_wnd)
```

```python
df_promoter_wnd["windowstart"]=df_promoter_wnd["start"]+df_promoter_wnd["indexarray"]
df_promoter_wnd["windowend"]=df_promoter_wnd["windowstart"]+201
df_promoter_wnd
```

```python
df_promoter_cod=df_promoter_wnd.drop(['indexarray','start','end'], axis=1)


df_promoter_cod_rearg=df_promoter_cod[["chrom", "windowstart", "windowend","gene"]]
df_promoter_cod_rearg
```

```python
import pybedtools

fbed = pybedtools.BedTool.from_dataframe(df_promoter_cod_rearg)\
              .saveas('soy_WKRY_DEG5999_target_windows_201bp_0.8thresh_10stride.bed')
```
